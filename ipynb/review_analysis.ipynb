## Textual Analysis

```python
# load libraries
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
import sys  
!pip install contractions
import contractions
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
#initialising stopwords
all_stopwords = stopwords.words('english')
```

```python
#create POS tagging and initialise lemmatizer
def get_wordnet_pos(word):
    """Maps two-letter POS tag to a WordNet POS tag"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN) #if not one of four tags, default to noun tag

lemmatizer = WordNetLemmatizer()
```

```python
cleaned_text = [] #creating empty list to store cleaned reviews
for review in reviews.iterrows():
    sent= review[1]['body']
    sent= sent.lower() #convert to lowercase
    sent= contractions.fix(sent) #expands contractions
    sent= re.sub(r'[^a-zA-Z0-9]', ' ', sent) #removes punctuation and non-alphabetic characters, substitutes with space 
    
    words=nltk.word_tokenize(sent) #tokenizes string to a bag of words
    words= [i for i in words if not i in all_stopwords] #remove stopwords
    words= [lemmatizer.lemmatize(word,get_wordnet_pos(word)) for word in words] #lemmatize 
    sent= ' '.join(words) #join back 
    cleaned_text.append(sent) # append the cleaned text to the list
    
reviews['cleaned_text'] = cleaned_text
```

## Topic modelling

```python
from sklearn.decomposition import LatentDirichletAllocation
import matplotlib.pyplot as plt
```

```python
def plot_top_words(model, feature_names, n_top_words, title, rows=2, columns=5):
    fig, axes = plt.subplots(rows, columns, figsize=(30, 15), sharex=True)
    axes = axes.flatten()
    for topic_ind, topic in enumerate(model.components_):
        top_features_ind = topic.argsort()[:-n_top_words - 1:-1]
        top_features = [feature_names[i] for i in top_features_ind]
        weights = topic[top_features_ind] #gives weight of each top feature

        ax = axes[topic_ind]
        ax.barh(top_features, weights, height=0.7,color='slateblue')
        ax.set_title(f'Topic {topic_ind +1}',
                     fontdict={'fontsize': 30})
        ax.invert_yaxis()
        ax.tick_params(axis='both', which='major', labelsize=20)
        for i in 'top right left'.split():
            ax.spines[i].set_visible(False)
        fig.suptitle(title, fontsize=40)

    plt.subplots_adjust(top=0.90, bottom=0.05, wspace=0.90, hspace=0.3)
    plt.savefig('topics', dpi=300, bbox_inches='tight')
    plt.show()
```

```python
corpus=reviews['cleaned_text'].tolist()
```

```python
# tokenization
no_features = 500 #gets top 500 features
vectorizer = CountVectorizer(analyzer='word',max_df=0.95, min_df=2,max_features=no_features)
X = vectorizer.fit_transform(corpus) #transform corpus into a sparse matrix -- bag of words 

tf_feature_names = vectorizer.get_feature_names_out()
```

```python
#run LDA
no_topics=3
lda = LatentDirichletAllocation(no_topics, max_iter=3, learning_method='online', learning_offset=50.,random_state=0).fit(X)
```

```python
# plot top words
no_top_words = 10
plot_top_words(lda, tf_feature_names, no_top_words, 'Topics in LDA model')
```
